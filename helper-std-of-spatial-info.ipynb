```python
# just certain sessions
df_filt = dfr.query("session_day in ['3_3','1_4']").copy()

# OPTION: non reward day 3 remove 
#df_filt = df_filt[~df_filt.eval("reward=='Out' and session_day =='3_3'")]
```

```python
#key="spatial_information"
key="sparsity"


#get repeats
df_filt = df_filt[["grp","roi_label","session_day",key]].reset_index()

roi_repeated = df_filt.groupby("roi_label").count().query("index==2").index

df_repeats = df_filt.query("roi_label in @roi_repeated").reset_index()
df_repeats = df_repeats.drop(columns=["level_0","index"])
```

```python
# pivot to calc diff
pivot = df_repeats.pivot_table(index=["grp","roi_label"],columns="session_day",values=key)
pivot.reset_index(inplace=True)
pivot["diff"] = pivot["1_4"] - pivot["3_3"]
pivot
```

```python
sns.boxplot(data=pivot, x = "grp",y="diff",showfliers=False)
```

```python
# see box
pivot.groupby("grp").boxplot(column="diff")
```

```python
pg.plot_paired(data=pivot.reset_index(), dv='spatial_information', within='session_day',
                    subject='roi_label')
```
