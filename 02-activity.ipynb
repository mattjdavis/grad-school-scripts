```python
def stim_response_summary_grouped(
        exptGrps, stim,ax, pre_time=None, post_time=None, channel='Ch2',
        label=None, roi_filters=None,colors=None,norm_pre=False):

    
    for expt_grp,color,roi_filter in zip(exptGrps,colofrom scipy.stats import spearmanr, pearsonr, linregressrs,roi_filters):
        normalized_psths,min_x_range=ia.PSTH(expt_grp, stim, pre_time=pre_time, post_time=post_time,
                                 channel=channel, label=label,return_full='norm',roi_filter=roi_filter)
        #mean
        mean_psth = np.nanmean(normalized_psths, axis=0)
        
        #sem
        ste_psth = np.nanstd(normalized_psths, axis=0) / np.sqrt(
                    normalized_psths.shape[0])
        
        if norm_pre:
            min_psth=np.min(mean_psth)
            mean_psth-=min_psth
            #ste_psth-=min_psth
        
        ax.plot(min_x_range, mean_psth,
                color='k' if color is None else color, lw=2)
        ax.fill_between(
                min_x_range, mean_psth - ste_psth, mean_psth + ste_psth,
                facecolor='r' if color is None else color, alpha=0.4)
        
    ax.axvline(0, linestyle='dashed', color='k')
    ax.set_xlim(min_x_range[0], min_x_range[-1])

    ax.set_xlabel('Time (s)')
    ax.set_ylabel(r'Mean $\Delta$F/F')
        
    #ax.set_title(stim)
    #ax.set_xlabel('')
    #ax.tick_params(axis='x', labelbottom=False)

    return normalized_psths,min_x_range
```
